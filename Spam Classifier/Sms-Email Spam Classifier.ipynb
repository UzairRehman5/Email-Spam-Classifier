# Text Preprocessing

```python
def transform_text(text):
    # Lower casing
    text = text.lower()

    # Tokenization
    text = nltk.word_tokenize(text)

    # Removing HTML tags
    text = [BeautifulSoup(word, 'html.parser').get_text() for word in text]

    # Removing special characters
    text = [i for i in text if i.isalnum()]

    # Removing stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    text = [i for i in text if i not in stop_words and i not in string.punctuation]

    # Stemming
    ps = PorterStemmer()
    text = [ps.stem(i) for i in text]

    return " ".join(text)
```

```python
df['transform_text'] = df['text'].apply(transform_text)
```

# Text Vectorization

- **Bag of Words**

```python
cv_bow = CountVectorizer()

X = cv_bow.fit_transform(df['transform_text']).toarray()
y = df['target'].values
```

```python
X_bow_train,X_bow_test,y_bow_train,y_bow_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

- **N-grams**

```python
cv_n_grams = CountVectorizer(ngram_range=(1,3))

X = cv_n_grams.fit_transform(df['transform_text']).toarray()
y = df['target'].values
```

```python
X_ngrams_train,X_ngrams_test,y_ngrams_train,y_ngrams_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

- **Tf-Idf**

```python
tfidf = TfidfVectorizer()

X = tfidf.fit_transform(df['transform_text']).toarray()
y = df['target'].values
```

```python
X_tf_train,X_tf_test,y_tf_train,y_tf_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

# Modelling

```python
svc = SVC()
knc = KNeighborsClassifier()
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
dtc = DecisionTreeClassifier(max_depth=5)
lrc = LogisticRegression()
rfc = RandomForestClassifier(n_estimators=50, random_state=2)
xgb = XGBClassifier(n_estimators=50,random_state=2)
```

```python
clfs = {
    'Support Vector' : svc,
    'K-Neighbors' : knc,
    'Gaussian_NB' : gnb,
    'Multinomial_NB' : mnb,
    'Bernoulli_NB' : bnb,
    'Decision Tree' : dtc,
    'Logistic Regression' : lrc,
    'Random Forest' : rfc,
    'XG Boost' : xgb
}
```

```python
def train_classifier(clf,X_train,y_train,X_test,y_test):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    
    return accuracy,precision
```

- **Bag of Words ---> Machine Learning**

```python
accuracy_scores = []
precision_scores = []

for name,clf in clfs.items():
    
    current_accuracy,current_precision = train_classifier(clf, X_bow_train, y_bow_train, X_bow_test, y_bow_test)
    
    print("For ",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)
    print(' ')
    
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
```

```python
performance_bow_df = pd.DataFrame({'Algorithm(BOW)':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores}).sort_values('Precision',ascending=False)
performance_bow_df
```

- **N-grams ---> Machine Learning**

```python
accuracy_scores = []
precision_scores = []

for name,clf in clfs.items():
    
    current_accuracy,current_precision = train_classifier(clf, X_ngrams_train, y_ngrams_train, X_ngrams_test, y_ngrams_test)
    
    print("For",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)
    print(' ')
    
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
```

```python
performance_ngrams_df = pd.DataFrame({'Algorithm(N-grams)':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores}).sort_values('Precision',ascending=False)
performance_ngrams_df
```

- **Tf-Idf ---> Machine Learning**

```python
accuracy_scores = []
precision_scores = []

for name,clf in clfs.items():
    
    current_accuracy,current_precision = train_classifier(clf, X_tf_train, y_tf_train, X_tf_test, y_tf_test)
    
    print("For",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)
    print(' ')
    
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
```

```python
performance_tfidf_df = pd.DataFrame({'Algorithm(Tf-Idf)':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores}).sort_values('Precision',ascending=False)
performance_tfidf_df
```
